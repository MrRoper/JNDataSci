```python
#import matplotlib.pyplot as plt
#import seaborn as sns
import pandas as pd
from bokeh.charts import BoxPlot, Scatter, show

auto = pd.read_csv("Automobile_price_data _clean.csv")
autoND = pd.read_csv("Automobile_price_data _NoDash.csv")
```

```python
auto.head()
```

```python
autoND.head()
```

```python
p = BoxPlot(auto, values = "price", label = ["make", "body-style"], color = "make", plot_width = 1000, plot_height = 800, title = "Price Summary (grouped by make,body-style)")
show(p)
```

```python
tooltips=[
  ('Price', '@price'),
  ('Make', '@make'),
  ('Body-style', '@bodystyle'),
  ('City-mpg', '@citympg'),    
]

p2 = Scatter(autoND, x = "citympg", y = "price", color = "make", marker = "bodystyle", title = "LAS350A - Lab 10: Price vs City-mpg (shaded by Make, maker by Body-style)", plot_width = 1000, plot_height = 800, tooltips=tooltips)
show(p2)
```
