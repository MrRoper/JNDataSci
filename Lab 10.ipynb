```python
import pandas as pd
from bokeh.charts import BoxPlot, Scatter, show

auto = pd.read_csv("Automobile_price_data _clean.csv")
autoND = pd.read_csv("Automobile_price_data _NoDash.csv")
```

```python
auto.head()
```

```python
autoND.head()
```

```python
p = BoxPlot(auto, values = "price", label = ["make", "body-style"], color = "make", plot_width = 800, plot_height = 600, title = "Price Summary (grouped by make,body-style)")
show(p)
```

The above line is supposed to produce a graph but GitHub is not able to show it. These differ from seaborn and matplotlib graphs because they are interactive. 
You can view the graph here: http://htmlpreview.github.io/?https://github.com/MrRoper/JNDataSci/blob/master/Lab10%20First%20Graph.html

```python
tooltips=[
  ('Price', '@price'),
  ('Make', '@make'),
  ('Body-style', '@bodystyle'),
  ('City-mpg', '@citympg'),    
]

p2 = Scatter(autoND, x = "citympg", y = "price", color = "make", marker = "bodystyle", title = "LAS350A - Lab 10: Price vs City-mpg (shaded by Make, maker by Body-style)", plot_width = 800, plot_height = 600, tooltips=tooltips)
show(p2)
```

Same thing as the first one. The only difference is this one has tooltips.
